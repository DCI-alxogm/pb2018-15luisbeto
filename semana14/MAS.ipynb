## MOVIMIENTO ARMONICO SIMPLE

#### DECLARACIÓN DE LIBRERIAS

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### DECLARACIÓN DE FUNCIONES

```python
## Formula para calcular la velucidad
def velocidad(v,h,T,O):
    return v+h*(-((2*np.pi/(T))**2)*O)
## Formula para calcular el  angulo
def angulo(O,h,v):
    return O+h*v
```

#### DECLARACIÓN DE VARIABLES

```python
## "u" variable de repeticiones
u=100000
## "vt" vector de velocidad
vt=np.linspace(0,10,u)
## "Ot" vector de angulo
Ot=np.linspace(180,10,u)
## "t" vector de tiempo
t=np.linspace(0,10,u)
## "T" variable de periodo
T=100
## "h" variable de incremento
h=0.005
## "w" variable de velocidad angular
w=2*np.pi/(T)
```

#### CALCULOS DE LA VELOCIDAD Y EL ANGULO CONRESPECTO DE EL TIEMPO

```python
for i in range (u-1):
    ## Operación para asignar los valores de las velocidades después de cada incrimento
    vt[i+1]=velocidad(vt[i],h,T,Ot[i])
    ## Operación para asignar los valores de los angulos después de cada incremento
    Ot[i+1]=angulo(Ot[i],h,vt[i])
```

#### GRÁFICA DE VELOCIDAD VS ANGULO

```python
plt.plot(vt,Ot, color = 'blue', label = 'Velocidad VS Angulo')
plt.xlabel('Velocidad')
plt.ylabel('Angulo')
plt.legend()
```

#### GRÁFICA DE TIEMPO VS VELOCIDAD

```python
plt.plot(t,vt, color = 'blue', label = 'Tiempo VS Velocidad')
plt.xlabel('Tiempo')
plt.ylabel('Velocidad')
plt.legend()
```

#### GRÁFICA DE TIEMPO VS ANGULO 

```python
plt.plot (t,Ot, color = 'blue', label = 'Tiempo VS Angulo')
plt.xlabel('Tiempo')
plt.ylabel('Angulo')
plt.legend()
```
